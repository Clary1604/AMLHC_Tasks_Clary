```python
import pandas as pd # für datentyp dataframe
import math as mt
import matplotlib.pyplot as plt # zum visualisieren von daten
import numpy as np # für datentyp array, funktionen, matrizen
from sklearn import datasets # für iris datensatz
```

```python
# creata a calcMeans function (berechnet Mittelwert für jede Spalte im Dataframe)
def calcMeans(dataframe):
    return dataframe.mean()


# use apply function: (geht aber auch so wie oben) 
def calcMeans2(dataframe: pd.DataFrame) -> pd.DataFrame:
    return dataframe.apply(np.mean)


# use describe function
```

```python
# call calcMeans function
iris_X, iris_y = datasets.load_iris(return_X_y=True, as_frame=True) # lädt den iris datensatz in die variable iris_X und iris_y

#print(iris_X) # ist array mit den 4 messungen der blumen
#print(iris_y) # ist Vektor, welche Blumenart es ist, d.h. label 
print("calcMeans:")
calcMeans(iris_X)

```

```python
print("calcMeans2:")
calcMeans2(iris_X)
```
